#### Importing Libraries

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import json
import random
import time
from time import sleep
```

#### Loading the Url we want to Scrape

```python
url = 'https://store.steampowered.com/search/results/?query&start=0&count=50&dynamic_data=&force_infinite=1&ndl=1&snr=1_7_7_230_7&infinite=1'
```

#### Checking response and total elements count

```python
def totalresults(url):
    r = requests.get(url)
    print(r)
    data = dict(r.json())
    totalresults = data['total_count']
    return int(totalresults)
```

```python
totalresults(url)
```

#### Defining a function to get all the data from the URL

```python
def get_data(url):
    r = requests.get(url)
    data = dict(r.json())
    return data['results_html']
```

```python
print(get_data(url))
```

#### Defining a function to parse the data

```python
def parse(data):
    gameslist = []
    soup = BeautifulSoup(data, 'html.parser')
    games = soup.find_all('a')
    for game in games:
        title = game.find('span', {'class': 'title'}).text
        price = game.find('div', {'class': 'search_price'}).text.strip().split('€')[0]
        try:
            discprice = game.find('div', {'class': 'search_price'}).text.strip().split('€')[1]
        except:
             discprice = price
            
#         print(title,price,discprice)
        
        mygame = {
            'title': title,
            'price': price,
            'discprice': discprice
        }
        gameslist.append(mygame)
    return gameslist
```

#### Building the Loop to go through all the elements

```python
%%time
results = []
for x in range(0,totalresults(url),50):
    wait_time = random.randint(1,4)
    data = get_data(f"https://store.steampowered.com/search/results/?query&start={x}&count=50&dynamic_data=&force_infinite=1&ndl=1&snr=1_7_7_230_7&infinite=1")
    results.append(parse(data))
    print('Napping for ',wait_time,'s')
    print('Results Scraped: ', x)
    time.sleep(wait_time)
```

#### Defining a function to export the data as csv.

```python
def output(results):
    gamesdf = pd.concat([pd.DataFrame(g) for g in results])
    gamesdf.to_csv('gamesprices2.csv',index=False)
    return
```

```python
output(results)
```
